## Album data

```python
album_list = []
for row in spotify_data["items"]:
    album_id = row["track"]["album"]["id"]
    album_name = row["track"]["album"]["name"]
    album_release_date = row["track"]["album"]["release_date"]
    album_total_tracks = row["track"]["album"]["total_tracks"]
    album_urls = row["track"]["album"]["external_urls"]["spotify"]
    album_element = {'album_id' : album_id, 'album_name': album_name, 'release_date' :album_release_date,
                     'total_tracks' : album_total_tracks , 'album_urls' : album_urls}

    album_list.append(album_element)


    
```

```python
album_df = pd.DataFrame(album_list)
album_df.head()
```

```python
album_df = album_df.drop_duplicates(subset="album_id")
```

```python
album_df["release_date"] = pd.to_datetime(album_df["release_date"])
```

## Artist Data

```python
artist_list = []

for row in spotify_data["items"]:
    for key, value in row.items():
        if key == "track":
            # print(value["artists"])
            for artist in value["artists"]:
                artist_id = artist["id"]
                artist_name = artist["name"]
                external_url = artist["href"]
                
                artist_element = {"Artist_id" : artist_id, "Name" : artist_name, "external_url" : external_url}
                artist_list.append(artist_element)

```

```python
artists_df = pd.DataFrame(artist_list)
artists_df.head()
```

```python
artists_df = artists_df.drop_duplicates(subset="Artist_id")
```

## Song Data

```python
song_list = []
for row in spotify_data["items"]:
    song_id = row["track"]["id"]
    song_name = row["track"]["name"]
    song_duration = row["track"]["duration_ms"]
    song_url = row["track"]["external_urls"]["spotify"]
    song_popularity = row["track"]["popularity"]
    song_added = row["added_at"]
    album_id = row["track"]["album"]["id"]
    artist_id = row["track"]["album"]["artists"][0]['id']

    song_element = {'song_id' : song_id, 'song_name': song_name, 'song_duration' :song_duration,
                     'song_url' : song_url , 'song_popularity' : song_popularity, 'song_added': song_added,'album_id': album_id,
                   'artist_id': artist_id}

    song_list.append(song_element)
```

```python
song_df = pd.DataFrame(song_list)
song_df.head()
```

```python
song_df = song_df.drop_duplicates(subset="song_id")
```

```python
song_df["song_added"] = pd.to_datetime(song_df["song_added"])
```
